```python
import cv2
import imghdr
from matplotlib import pyplot as plt
```

```python
pikachu_data = 'pikachu'
```

```python
image_exts = ['jpeg','jpg','bmp','png']
```

```python
os.listdir(pikachu_data)
```

```python
img = cv2.imread(os.path.join('pikachu','image31.jpeg'))
```

```python
for image in os.listdir(pikachu_data):
    img_path = os.path.join(pikachu_data, image)
    try:
        img = cv2.imread(img_path)
        tip = imghdr.what(img_path)
        if tip not in image_exts:
            print('Obrazek ma nespravny format {}'.format(img_path))
            os.remove(img_path)
    except Exception as e:
            print('Issue with image{}'.format(img_path))
```
