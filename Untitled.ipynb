```python
 for dir_number in os.listdir(data_slozka): 
    for image in os.listdir(os.path.join(data_slozka, dir_number)):
        img_path = os.path.join(data_slozka, dir_number, image)
        try:
            img = cv2.imread(img_path)
            tip = imghdr.what(img_path)
            if tip not in image_exts:
                print('Obrazek ma nespravny format {}'.format(img_path))
                os.remove(img_path)
        except Exception as e:
                print('Issue with image{}'.format(img_path))
```
